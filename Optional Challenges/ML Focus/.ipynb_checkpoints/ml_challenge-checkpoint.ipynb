## Training

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_, Y, test_size=.2, stratify=Y)
```

```python
unique, counts = np.unique(y_train, return_counts=True)
print(np.array((unique, counts)))
```

```python
rfc = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
rfc.fit(X_train, y_train)
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 

svc = GridSearchCV( SVC(), param_grid, refit = True, scoring = 'accuracy')
```

```python
svc.fit(X_train, y_train)
```

## Testing / Optimization

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.colors import ListedColormap
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
cm = plt.cm.RdBu
cm_bright = ListedColormap(["#FF0000", "#0000FF"])


#do the pretty colors
display = DecisionBoundaryDisplay.from_estimator(
    rfc, X_, cmap=cm, alpha=0.8
)

# Plot the training points
display.ax_.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, alpha=0.2, edgecolors="k")
# Plot the testing points
display.ax_.scatter(
    X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=1, edgecolors="k"
)

plt.title('Random Forest Classifier')
plt.show()
```

```python
cm = plt.cm.RdBu
cm_bright = ListedColormap(["#FF0000", "#0000FF"])


#do the pretty colors
display = DecisionBoundaryDisplay.from_estimator(
    svc, X_, cmap=cm, alpha=0.8
)

# Plot the training points
display.ax_.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, alpha=0.8, edgecolors="k")
# Plot the testing points
display.ax_.scatter(
    X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=.4, edgecolors="k"
)

plt.title('SVC')
plt.show()
```

## Performance Analysis

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

random forest classifier:

```python
y_rfc_predict = rfc.predict(X_test)
```

```python
print(rfc.score(X_train, y_train))
print(rfc.score(X_test, y_test))
```

```python
confusion_matrix(y_test, y_rfc_predict)
# true negative        false positive
# false negative       true postive
```

```python
print(classification_report(y_test, y_rfc_predict))
```

SVC:

```python
y_svc_predict = svc.predict(X_test)
```

```python
print(svc.score(X_train, y_train))
print(svc.score(X_test, y_test))
```

```python
confusion_matrix(y_test, y_svc_predict)
# true negative        false positive
# false negative       true postive
```

```python
print(classification_report(y_test, y_svc_predict))
```
